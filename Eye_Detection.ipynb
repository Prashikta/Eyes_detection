# Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Open Image

```python
img = cv2.imread('Images/aish.jpeg')
```

```python
plt.imshow(img)
```

# Fixing the color of image

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

# Create Classifier

```python
eye_classifier = cv2.CascadeClassifier('anaconda3\Lib\site-packages\cv2\data\haarcascade_eye.xml')
```

# Create a function to detect eyes

```python
def detect_eye(fix_img):
    eye_rects = eye_classifier.detectMultiScale(fix_img)
    
    for (x, y, w, h) in eye_rects:
        cv2.rectangle(fix_img, (x, y), (x+w, y+h), (255,255,255), 10)
    return fix_img
```

# Result

```python
result = detect_eye(fix_img)
plt.imshow(result)
```
